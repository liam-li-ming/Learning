```python
top_level30 = data30[data30['IVRS LEVEL'] == 2]['Column Name']
levels = {'level2': [], 'level3': [], 'level4': []}

for child in top_level30:
    level2 = child
    level3 = data30.loc[data30['Column Group'] == level2, 'Column Name'].tolist()
    
    if level3:
        for l3 in level3:
            level4 = data30.loc[data30['Column Group'] == l3, 'Column Name'].tolist()
            if level4:
                for l4 in level4:
                    levels['level2'].append(level2)
                    levels['level3'].append(l3)
                    levels['level4'].append(l4)
            else:
                levels['level2'].append(level2)
                levels['level3'].append(l3)
                levels['level4'].append(None)
    else:
        levels['level2'].append(level2)
        levels['level3'].append(None)
        levels['level4'].append(None)

result_df = pd.DataFrame(levels)
result_df

result_df.insert(0, 'Hotline', 1830830)
result_df
```

```python
def expected_share_price(Dt, gs, gl, n, r): 
    sum_short = 0
    for i in range(1, n + 1):
        sum_short += (Dt * (1 + gs) ** i) / ((1 + r)**i)
        print(sum_short)
    E_Stn = ((Dt * ((1 + gs)**n) * (1 + gl))/(r - gl)) * (1 + r)**(-n)
    print(E_Stn)

    print("Result", sum_short + E_Stn)
```
