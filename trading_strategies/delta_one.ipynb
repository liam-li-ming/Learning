#### ETF Arbitrage

#### To calculate the fair value of an ETF

```python
def calculate_fair_value(composition_data, market_prices, management_fee, accrued_dividends):
    """
    This uses YOUR data to calculate what ETF should trade at
    """
    nav = 0
    
    for component in composition_data:
        ticker = component['ticker']
        weight = component['weight']
        market_price = market_prices[ticker]
        
        nav += weight * market_price
    
    # Adjust for fees, dividends, etc.
    nav = nav - management_fee
    nav = nav + accrued_dividends
    
    return nav
```
